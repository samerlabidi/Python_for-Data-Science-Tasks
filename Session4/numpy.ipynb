#### 5. Create a 5x2x3 3-dimensional array with all values equaling 1. Assign the array to variable *b*.

```python
b = np.ones((5, 2, 3))
print(b.shape)
```

#### 6. Print *b*.

```python
print(b)
```

#### 7. Do *a* and *b* have the same size? How do you prove that in Python code?

```python
print('a.size =', a.size)
print('b.size =', b.size)
print('Same size?', a.size == b.size)
# Yes, both have size 30 (2*3*5 = 5*2*3 = 30), even though shapes differ.
```

#### 8. Are you able to add *a* and *b*? Why or why not?

```python
try:
    print(a + b)
except ValueError as e:
    print(type(e).__name__ + ':', e)

# Answer: No. a has shape (2, 3, 5) and b has shape (5, 2, 3).
# NumPy cannot broadcast these shapes for element-wise addition.
```

#### 9. Transpose *b* so that it has the same structure of *a* (i.e. become a 2x3x5 array). Assign the transposed array to variable *c*.

```python
# b is (5, 2, 3). We need (2, 3, 5).
c = np.transpose(b, (1, 2, 0))
print(c.shape)
```

#### 10. Try to add *a* and *c*. Now it should work. Assign the sum to variable *d*. But why does it work now?

```python
d = a + c
print(d.shape)
# It works because a and c now have the same shape (2, 3, 5),
# so element-wise addition is valid.
```

#### 11. Print *a* and *d*. Notice the difference and relation of the two array in terms of the values? Explain.

```python
print('a:\n', a)
print('\nd:\n', d)
# Relation: since c is all ones, d = a + 1 for every element.
# Each value in d is exactly 1 greater than the corresponding value in a.
```

#### 12. Multiply *a* and *c*. Assign the result to *e*.

```python
e = a * c
print(e)
```

#### 13. Does *e* equal to *a*? Why or why not?

```python
print(np.array_equal(e, a))
print(np.allclose(e, a))
# Yes. c is an array of ones with the same shape as a,
# so multiplying a * c leaves every value unchanged. Therefore e equals a.
```

#### 14. Identify the max, min, and mean values in *d*. Assign those values to variables *d_max*, *d_min* and *d_mean*.

```python
d_max = d.max()
d_min = d.min()
d_mean = d.mean()
print(d_max, d_min, d_mean)
```

#### 15. Now we want to label the values in *d*. First create an empty array *f* with the same shape (i.e. 2x3x5) as *d* using `np.empty`.

```python
f = np.empty(d.shape)
print(f.shape)
```

#### 16. Populate the values in *f*.

For each value in *d*, if it's larger than *d_min* but smaller than *d_mean*, assign 25 to the corresponding value in *f*. If a value in *d* is larger than *d_mean* but smaller than *d_max*, assign 75 to the corresponding value in *f*. If a value equals to *d_mean*, assign 50 to the corresponding value in *f*. Assign 0 to the corresponding value(s) in *f* for *d_min* in *d*. Assign 100 to the corresponding value(s) in *f* for *d_max* in *d*. In the end, f should have only the following values: 0, 25, 50, 75, and 100.

```python
for i in range(d.shape[0]):
    for j in range(d.shape[1]):
        for k in range(d.shape[2]):
            value = d[i, j, k]
            if value == d_min:
                f[i, j, k] = 0
            elif value == d_max:
                f[i, j, k] = 100
            elif value == d_mean:
                f[i, j, k] = 50
            elif d_min < value < d_mean:
                f[i, j, k] = 25
            elif d_mean < value < d_max:
                f[i, j, k] = 75

print(np.unique(f))
```

#### 17. Print *d* and *f*. Do you have your expected *f*?

```python
print('d:\n', d)
print('\nf:\n', f)
```

#### 18. Bonus question: instead of using numbers (i.e. 0, 25, 50, 75, and 100), use string values  ("A", "B", "C", "D", and "E") to label the array elements.

```python
g = np.empty(d.shape, dtype=object)

for i in range(d.shape[0]):
    for j in range(d.shape[1]):
        for k in range(d.shape[2]):
            value = d[i, j, k]
            if value == d_min:
                g[i, j, k] = 'A'
            elif value == d_max:
                g[i, j, k] = 'E'
            elif value == d_mean:
                g[i, j, k] = 'C'
            elif d_min < value < d_mean:
                g[i, j, k] = 'B'
            elif d_mean < value < d_max:
                g[i, j, k] = 'D'

print(g)
```
