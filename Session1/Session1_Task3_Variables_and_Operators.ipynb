1.1 For the following variables, use the function type() from Python to check the data type for the variables.

```python
x1 = 1.1
x2 = "Monday"
x3 = "1.1"
x4 = True
x5 = "True"
x6 = -1
```

```python
print(type(x1))
print(type(x2))
print(type(x3))
print(type(x4))
print(type(x5))
print(type(x6))
```

1.2 What is the difference between variables x1 and x3?

```python
print(type(x1))
print(type(x3))
# Conclusion: x1 is a float (numeric 1.1), while x3 is a string containing the characters '1.1'.
```

1.3 What happens if you subtract x3 from x1?

```python
try:
    result = x1 - x3
    print(result)
except TypeError as e:
    print(type(e).__name__ + ':', e)
# Conclusion: You get a TypeError because you cannot subtract a string from a float.
```

1.4 What is the difference between variables x4 and x5?

```python
print(type(x4))
print(type(x5))
# Conclusion: x4 is a boolean (True), while x5 is a string containing the text 'True'.
```

1.5 What happens if you subtract x5 from x4?

```python
try:
    result = x4 - x5
    print(result)
except TypeError as e:
    print(type(e).__name__ + ':', e)
# Conclusion: You get a TypeError because you cannot subtract a string from a bool.
```

Question 2: What is the data type of x1 and x2? Now perform simple comparisons between x1 and x2:

Check if the two variables are equal?

Check if x1 is greater than x2?

Check if x2 is greater than x1?

Check if x1 is not equal to x2?

Store the difference between the two variables x1 and x2 in another variable x3 (subtract the smaller number from the larger number).

Increment the smaller of the two variables (x1 and x2) with the difference. Use the shorthand addition operator for the same. Again check if x1 and x2 are equal or not?

```python
print(type(x1))
print(type(x2))

print('Equal?', x1 == x2)
print('x1 > x2?', x1 > x2)
print('x2 > x1?', x2 > x1)
print('Not equal?', x1 != x2)

if x1 >= x2:
    x3 = x1 - x2
    smaller = 'x2'
else:
    x3 = x2 - x1
    smaller = 'x1'

print('Difference x3:', x3)

if smaller == 'x1':
    x1 += x3
else:
    x2 += x3

print('After increment: x1 =', x1, ', x2 =', x2)
print('Equal now?', x1 == x2)
# Conclusion: after adding the difference to the smaller value, both variables become equal.
```
