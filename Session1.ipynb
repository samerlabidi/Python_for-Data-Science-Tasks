```python
import math

def sqrList(l):
  S=[]
  for i in range(len(l)):
    S.append(math.sqrt(l[i]))
  return S
```

```python
sqrList([1,2,3,4])
```

```python
def strlink (S1,S2):
  S3=""
  len1 = len(S1)
  len2 = len(S2)

  min_len = min(len1, len2)

  for i in range(min_len):
    S3 += S1[i]
    S3 += S2[len2 - 1 - i]

  if len1 > min_len:
    S3 += S1[min_len:]
  elif len2 > min_len:
    S3 += S2[:len2 - min_len]

  return S3
```
